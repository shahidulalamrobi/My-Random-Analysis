```python
import pandas as pd
import numpy as np
```

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot') 
```

```python
bn=pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
```

```python
bn.head()
```

```python
df=bn.dropna(axis=0)
```

```python
df.head()
```

```python
df2=df[df.location=='Bangladesh']
```

```python
df2.head()
```

```python
df3['total_cases'].plot(figsize=(15,8))
plt.title('Total Cases Bangladesh to Date')
plt.show()
```

```python
df3=df2.set_index('date')
```

```python
df3['new_cases'].plot(figsize=(15,8))
plt.title('New Cases Bangladesh to Date')
```

```python
df2
```

```python
df3.head()
```

```python
df3['new_cases'].plot(figsize=(15,8))
plt.title('New Cases Bangladesh to Date')
```

```python
df3[['total_cases','new_cases','new_deaths']].plot(kind='area',stacked=False,figsize=(15,8))
plt.title('Total, New Cases, and Deaths in Bangladesh to Date')
```

```python
df3['new_cases'].max()
```

```python
df3['total_cases'].plot(kind='hist',figsize=(15,8))
plt.title('Total Cases Bangladesh to Date in Histogram')
```

```python
dfn=(df3['total_cases'].shift(-1)/df3['total_cases']-1)*100
```

```python
dfn.plot(figsize=(15,8))
plt.title('Rate of Increase')
plt.show()
```

```python
df3['new_cases'].plot(kind='hist',figsize=(15,8))
plt.title('New Cases Bangladesh to Date in Hostogram')
```

```python
df3['total_cases'].plot(kind='bar',figsize=(15,8),color='red')
plt.title('Total Cases Bangladesh to Date in Bar')
```

```python
df3['new_cases'].plot(kind='bar',figsize=(15,8))
plt.title('Total Cases Bangladesh to Date in Bar')
```

```python
df3[['new_cases','total_cases']].plot(kind='scatter',x='new_cases',y='total_cases',figsize=(15,8),color='red')
plt.title('New Case &Total Case Relation')
```

```python
df3[['new_tests','new_cases']].plot(figsize=(15,8))
plt.title('New Test VS New Cases Bangladesh to Date')
```

```python
df3[['new_tests','new_cases']].plot(kind='scatter',x='new_tests',y='new_cases',
                                    figsize=(15,8),color='red')
plt.title('New Test VS New Cases Relation')
```
